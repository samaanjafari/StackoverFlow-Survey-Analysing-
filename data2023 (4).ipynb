```python
#There are couple of columns that aren't necessary that named knowledge_1,2 and ... and their existence just make our dataframe nasty so let's create a function that will do it for us.

def deleting_cols():
    cols = input('what are your string names that u want to erase from your df columns? ').split(' ')
    for i in cols:              
        for col in df.columns:
            if i in col:
                df.drop(col, axis=1 , inplace = True)
           
```
